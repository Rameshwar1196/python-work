```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
'''
** Create a figure object called fig using plt.figure() **
** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
** Plot (x,y) on that axes and set the labels and titles to match the plot below:**
'''
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('graph of x and y ')
```

```python
#* Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
#Now plot (x,y) on both axes. And call your figure object to show it.
ax1.plot(x,y)
ax1.set_xlabel('x-axis')
ax1.set_ylabel('y-axis')
ax1.set_title('outer graph')
ax2.plot(x,y)
ax2.set_xlabel('x-axis')
ax2.set_ylabel('y-axis')
ax2.set_title('inner graph')
fig
```

```python
# Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot
fig=plt.figure()
ax3=fig.add_axes([0,0,1,1])
ax3.plot(x,z)
ax3.set_xlabel('x')
ax3.set_ylabel('z')
ax3.set_title('outer graph')

ax4=fig.add_axes([0.2,0.5,0.2,0.2])
ax4.plot(x,y)
ax4.set_xlabel('x')
ax4.set_ylabel('y')
ax4.set_title('inner graph')
ax4.set_xlim(20,22)
ax4.set_ylim(35,50)
```

```python
#Use plt.subplots(nrows=1, ncols=2) to create the plot below
fig,axes =plt.subplots(nrows=1,ncols=2)
```

```python
#Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]
fig=plt.figure()
ax1=fig.add_axes( [0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])
```

```python
#Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:*
#fig=plt.figure()
fig,axes =plt.subplots(nrows=2,ncols=1)
ax1=fig.add_axes( [0,0,1,1])
ax1.plot(x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.set_title('outer graph')

ax2=fig.add_axes([0.2,0.5,.4,.4])
ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('inner graph')
ax2.set_xlim(20,22)
ax2.set_ylim(35,50)
fig
```

```python
# Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style
fig,ax=plt.subplots(nrows=1,ncols=2)
ax[0].plot(x,y,color="blue", lw=3, ls='--')
ax[1].plot(x,z,color="red", lw=3, ls='-')
```

```python
#See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.*
fig, ax= plt.subplots(nrows=1, ncols=2,figsize=(12,2))

ax[0].plot(x,y,color="blue", lw=5.5)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

ax[1].plot(x,z,color="red", lw=3, ls='--')
ax[1].set_xlabel('x')
ax[1].set_ylabel('z')
```
